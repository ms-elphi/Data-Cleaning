```python
def one_hot_top_x(df, variable, top_x_labels):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(data[variable]==label, 1, 0)

# read the data again
data = pd.read_csv("C:\\Users\\admin\PP_programs\DataRepo\mercedess\\train.csv ", usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# encode X2 into the 10 most frequent categories
one_hot_top_x(data, 'X2', top_10)
data.head()
```

```python
top_10 = [x for x in data.X1.value_counts().sort_values(ascending=False).head(10).index]

# now create the 10 most frequent dummy variables for X1
one_hot_top_x(data, 'X1', top_10)
data.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import roc_auc_score

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('C:\\Users\\admin\PP_programs\DataRepo\\titanic\\train.csv')
data.head()
```

```python
data_one_hot = pd.concat([data[['Pclass','Age','SibSp','Parch','Survived']],
                         pd.get_dummies(data.Sex,drop_first=True),
                         pd.get_dummies(data.Embarked,drop_first=True)],axis = 1)
```

```python
data_one_hot.head(5)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(data_one_hot[['Pclass','Age','SibSp','Parch','male','Q','S']].fillna(0),
                                                data_one_hot['Survived'],test_size=0.30,random_state=12)
```

```python
X_train.shape,X_test.shape
```

# Random Forest Classifier

```python
rf = RandomForestClassifier(n_estimators=100,max_depth=10,random_state=12)
rf.fit(X_train,Y_train)
pred = rf.predict_proba(X_train)
print('Train set: Random Forests roc-auc: {}'.format(roc_auc_score(Y_train, pred[:,1])))
pred = rf.predict_proba(X_test)
print('Test set: Random Forest roc-auc:{}'.format(roc_auc_score(Y_test,pred[:,1])))
```

# AdaBoost Classifier

```python
ada = AdaBoostClassifier(n_estimators=100,random_state=12)
ada.fit(X_train,Y_train)
pred = ada.predict_proba(X_train)
print('Train set: Ada Boost roc-auc: {}'.format(roc_auc_score(Y_train, pred[:,1])))
pred = ada.predict_proba(X_test)
print('Test set: Ada BOost roc-auc:{}'.format(roc_auc_score(Y_test,pred[:,1])))
```

# Logistic regression

```python
lreg = LogisticRegression(random_state=12,solver = 'lbfgs')
lreg.fit(X_train,Y_train)
pred = lreg.predict_proba(X_train)
print('Train set: Ada Boost roc-auc: {}'.format(roc_auc_score(Y_train, pred[:,1])))
pred = lreg.predict_proba(X_test)
print('Test set: Ada BOost roc-auc:{}'.format(roc_auc_score(Y_test,pred[:,1])))
```
